```python
data=pd.read_csv("final.csv")
data2=pd.read_csv("final_other.csv")
```

```python
data=data.drop(data[data.Gearbox.isna()].index)
data.reset_index(drop=True,inplace=True)
```

```python
data2=data2.drop(data2[data2.Hand_Drive.isna()].index)
data2.reset_index(drop=True,inplace=True)
```

```python
data2=data.drop(data[data["model"]=="Rare"].index)
data2.reset_index(drop=True,inplace=True)
data2.to_csv("final_without.csv",index_label=False)

reg=data2[["brand","model","year","Gearbox","Engine","price"]]

reg.dropna(inplace=True)

y=reg.price

x=reg[["brand","model","year","Gearbox","Engine"]]



x_dum=pd.get_dummies(x,drop_first=True)

test=data2[["brand","model","year","Gearbox","Engine","price"]]
test=test[test.price.isna()]
test.drop(columns="price",inplace=True)
dumm=pd.get_dummies(test,drop_first=True)
model = neighbors.KNeighborsRegressor(n_neighbors = 5)
x_train=x_dum[dumm.columns]
model.fit(x_train, y)  #fit the model
#pred=model.predict(x_test) #m

new_price=model.predict(dumm)

data2.loc[data2.price.isna(),"price"]=new_price
data2.info()

data.reset_index(drop=True,inplace=True)
data2.reset_index(drop=True,inplace=True)

data.to_csv("final.csv",index_label=False)
data2.to_csv("final_other.csv",index_label=False)

data.info()
```
